```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("titanic.csv", encoding="ISO-8859-1")
df.isnull().sum()
df.head()
df.info()
df.drop('Cabin',axis=1,inplace=True)
df["Age"].fillna(df["Age"].mean(), inplace=True)
df.head()
plt.title("Histogram of Ages")
plt.xlabel("Age")
df["Age"].plot.hist()
df.head()

```

```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
```

```python
plot_correlation_map( df )
```

```python
for i in range (0,len(df)):
    a=(df['Name'][i]).find(',')+2
    b=(df['Name'][i]).find('.') 
    df['Title'][i] = df['Name'][i][a:b]

df.head()
```

```python
df.head()
```

```python
for i in range (0,len(df)):
    if df['Title'][i] in ['Mr','Mrs','Miss','Ms','Mme','Mlle']:
        df['Title'][i]='Mrs'
    elif df['Title'][i] in ['Capt','Col','Major','Dr','Rev']:
        df['Title'][i]='Officer'
    elif df['Title'][i] in ['Jonkheer','Don','Sir','Lady','the Countess','Dona']:
        df['Title'][i]='Royalty'
   
df.head()
```

```python
g=sns.FacetGrid(df,col="Sex")
g.map(plt.hist,"Title",bins=10)
```
